+ pd.merge(team_data , player_data, on=index)

```python
import pandas as pd
import os
```

```python
def get_csv_files(path, league, data):
    csv_list = [file for file in os.listdir(path) if (file.startswith(league) and data in file)]
    return csv_list

team_csv_files = get_csv_files('./team csv files', 'LCK', 'team')
player_csv_files = get_csv_files('./player csv files', 'LCK', 'player')

team_csv_files, player_csv_files
```

```python
#리그, 연도가 같은 csv 파일 내 데이터 합병
for team_csv_file in team_csv_files:
    for player_csv_file in player_csv_files:
        if team_csv_file[:9] == player_csv_file[:9]:
            team_data =  pd.read_csv(f'./team csv files/{team_csv_file}')
            player_data = pd.read_csv(f'./player csv files/{player_csv_file}')
            
            merged_data = pd.merge(team_data, player_data, how='outer')
            merged_data.to_csv(f'{team_csv_file[:9]}data.csv')
```

```python
index = ['tournament name', 'date', 'version', 'match score', 'set', 'time']
pd.DataFrame(merged_data).set_index(index)
```
